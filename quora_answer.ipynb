```python
# Local input
f = open("input.txt", "r")
inp = f.readlines(1)[0]
n, m = tuple([int(x) for x in inp.strip().split(" ")])

df = pd.DataFrame()
X_train = []
y_train = []
for i in range(n):
    inp = f.readlines(1)[0]
    details = inp.split(" ")
    others = details[2:]
    y_train.append(int(details[1]))
    X_train.append([float(x.strip().split(":")[1]) for x in details[2:]])

X_train = np.array(X_train)
y_train = [1 if x > 0 else 0 for x in y_train]

inp = f.readlines(1)[0]
n_test = int(inp)

X_test = []
labels = []

test_df = pd.DataFrame()
for i in range(n_test):
    inp = f.readlines(1)[0]
    details = inp.split(" ")
    labels.append(details[0])
    X_test.append([float(x.strip().split(":")[1]) for x in details[1:]])

X_test = np.array(X_test)
X_test
```

```python
sg_clf = Pipeline([
    ('std', StandardScaler()),
    ('sgd', SGDClassifier(loss='modified_huber', penalty = 'l2', alpha = 0.1, learning_rate = 'optimal', max_iter = 500, random_state = 42))
])

sg_clf.fit(X_train, y_train)
sg_clf.score(X_train, y_train), cross_val_score(sg_clf, X_train, y_train, cv = 5)
```

```python
%%time
parameters = {'sgd__loss': ['hinge', 'log', 'modified_huber'],
             'sgd__penalty': ['l1','l2','elasticnet'],
             'sgd__alpha': [0.0001, 0.01, 0.1, 1],
              'sgd__learning_rate': ['optimal', 'adaptive'],
              'sgd__eta0': [0.001, 0.01, 0.1, 1]
             }

gs = GridSearchCV(sg_clf, parameters, cv=5)
gs.fit(X_train, y_train)
gs.best_score_, gs.best_params_
```

```python
rf_clf = Pipeline([
    ('std', StandardScaler()),
    ('rf', RandomForestClassifier(n_estimators = 50,
                                  max_depth = 100,
                                  min_samples_split = 2,
                                  min_samples_leaf = 1,
                                  random_state = 42))
])

rf_clf.fit(X_train, y_train)
rf_clf.score(X_train, y_train), cross_val_score(rf_clf, X_train, y_train, cv = 5)
```

```python
%%time
parameters = {'rf__n_estimators':[10, 50, 100, 200, 300],
             'rf__max_depth': [5, 10, 50, 100, 200],
             'rf__min_samples_split':[2, 5, 10, 50, 100],
             'rf__min_samples_leaf': [1, 5, 10, 50, 100]}

rs = RandomizedSearchCV(rf_clf, parameters, cv = 5)
rs.fit(X_train, y_train)
rs.best_score_, rs.best_params_
```

```python
results = np.where(sg_clf.predict(X_test) > 0, 1, -1)
for label, result in zip(labels, results):
    print("{:s} {:+d}".format(label, result))
```
